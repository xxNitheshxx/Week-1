```python
import pandas as pd
```

```python
df=pd.read_csv("/content/student-scores.csv")
df.columns
df.head()
```

```python
import matplotlib.pyplot as plt
```

```python
#Adding all the Scores to compare them
df['Total_Score']=df.iloc[:,10:17].sum(axis=1)
```

```python
#Generating Statistical data for the Numerical Columns
cols=df.describe().columns
means={}
median={}
mode={}
std_dev={}
for names in cols:
  if names=='id':
    continue
  means[names]=df[names].mean()
  median[names]=df[names].median()
  mode[names]=df[names].mode()
  std_dev[names]=df[names].std()
# print(means)
# print(median)
# print(mode)
# print(std_dev)
df_stat=pd.DataFrame(data=[means,median,mode,std_dev],index=['mean','median','mode','std'])
df_stat
```
